## 8.4 데이터 전처리

### 8.4.1 `etfTabName` 만들기 (`map()`, `apply()` 활용)

```python
etfcode = """전체
국내 시장지수
국내 업종/테마
국내 파생
해외 주식
원자재
채권
기타"""
etfcode
```

```python
etf_tab_name = etfcode.split('\n')
etf_tab_name
```

```python
def find_etf_tab_name(no):
    return etf_tab_name[no]
```

```python
find_etf_tab_name(1)
```

<br>

**`map()` vs `apply()`**

`map()`

- Series에서만 사용 가능
- 매개변수로 함수를 넣어주면 Series 값이 해당 함수의 인자로 들어가 결과가 반환됨
- 이름이 없는 함수를 사용할 때는 `lambda x: func(x)` 를 사용

```python
df['etfTabCode'].map(find_etf_tab_name)
```

```python
df['etfTabCode'].map(lambda x: etf_tab_name[x])
```

<br>

`apply()`

- Series와 DataFrame 둘 다 사용 가능
- 이름이 없는 함수를 사용할 때는 `lambda x: func(x)` 를 사용

```python
df['etfTabCode'].apply(find_etf_tab_name)
```

```python
df['etfTabCode'].apply(lambda x: etf_tab_name[x])
```

<br>

`etfTabName` 컬럼 생성

```python
df['etfTabName'] = df['etfTabCode'].apply(lambda x: etf_tab_name[x])
df.head()
```

```python
df.loc[df['etfTabCode'] == 1, ['etfTabCode', 'etfTabName', 'itemname']].head()
```

```python
df.loc[df['etfTabCode'] == 3, ['etfTabCode', 'etfTabName', 'itemname']].head()
```

### 8.4.3.2 인버스 여부

- 인버스 ETF는 지수가 하락하면 오히려 수익률이 오르도록 설계된 상품이다.

```python
df['인버스'] = df['종목명'].str.contains('인버스')
df.head()
```

```python
df['인버스'].value_counts()
```

<br>

#### 8.4.3.3 레버리지 여부

- 레버리지(leverage)는 타인의 자본을 지렛대처럼 이용하여 자기 자본의 이익률을 높이는 것이다.
- 고정비용이 있을 때 매출액의 변화가 기업의 손익에 미치는 영향을 분석하는 데 쓰인다.

```python
df['레버리지'] = df['종목명'].str.contains('레버리지')
df.head()
```

```python
df['레버리지'].value_counts()
```

```python
df['레버리지'].value_counts(normalize=True) * 100 # 비율 확인
```

<br>

#### 8.4.3.4 헤지(hedge)

- 환율, 금리 또는 다른 자산에 대한 투자 등을 통해 보유하고 있는 위험자산의 가격변동을 제거하는 것을 말한다.
- 즉, 확정되지 않은 자산을 확정된 자산으로 편입하는 과정이라 할 수 있다.
- 주로 선물 옵션과 같은 파생상품을 이용한다.
- 이를 통해 체계적 위험을 제거할 수 있다.  
  
  
- 완전헤지 : 부(wealth)를 결정하는 변수값의 변화와 관계없이 항상 일정한 부를 유지하게 하는 헤지
- 불완전헤지 : 부(wealth)를 결정하는 변수값의 변화에 따라 부가 변하는 헤지  
  
  
- 종목명에 포함된 `(H)` 의 의미는 환율변동 위험을 막기 위해 Hedge를 한다는 뜻이다.

```python
df['환헤지'] = df['종목명'].str.endswith('H)')
df.head()
```

```python
df[df['환헤지'] == True].head()
```

```python
df['환헤지'].value_counts()
```

```python
df['환헤지'].value_counts(normalize=True) * 100
```
