## 5.6 기간 수익률 비교하기

- 가격마다 스케일이 다르기 때문에 데이터프레임 기준 첫날을 0으로 맞추고 상대적으로 종가가 상승했는 지 하락했는 지에 대한 값을 구한다.
- 다른 스케일의 값을 상승/하락 비율을 비교해 볼 수 있다.  

<br>

### 5.6.1 다른 스케일의 값을 조정하는 방법

1. **표준화 (standardization)**
  - <font style='background-color:yellow; color: black'>데이터가 평균으로부터 얼마나 떨어져 있는 지</font>를 나타내는 값으로 변환
  - Z-score = (측정값 - 평균) / 표준편차  
  
  
2. **정규화 (normalization)**
  - <font style='background-color:yellow; color: black'>데이터의 상대적 크기에 대한 영향을 줄이기 위해</font> 0 ~ 1 사이의 값으로 변환
  - (측정값 - 최소값) / (최대값 - 최소값)

<br>

### 5.6.2 기간 수익률 구하기

```python
df2 = df[['LG화학', '삼성전자']]
df2
```

```python
df2.iloc[0]
```

```python
# 모든 날짜의 종가를 첫날 종가로 나눠준 후 1을 빼 준다.
df2_yield = df2 / df2.iloc[0] - 1
df2_yield
```

<br>

### 5.6.3 기간 수익률 시각화

```python
df2_yield.plot()
```

<br>

### 5.6.4 서브 플롯을 활용한 시각화

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

df[['삼성전자', 'LG화학']].plot(secondary_y='LG화학', ax=axes[0], title='종가 2축 그래프')
df2_yield.plot(ax=axes[1], title='수익률 그래프')
```

<br>

### 5.6.5 수익률 분포를 히스토그램으로 시각화

- 히스토그램 : 각 데이터의 빈도수 시각화

- LG화학이 삼성전자보다 수익률이 0보다 작은 기간이 더 적었다는 것을 알 수 있다.

## 5.7 왜도와 첨도

- 히스토그램의 분포를 추정해서 밀도추정곡선을 그리기도 한다.
- 여기에서는 왜도와 첨도를 구해 각 종목의 수익률 분포를 수치화된 데이터로 구해 본다.

<br>

### 5.7.1 왜도 (skewness, 비대칭도)

<img src="./img/05_01.jpg" style="width:600px; margin-left: 10px;" />

- 실수 값 확률 변수의 확률 분포 비대칭성을 나타내는 지표
- 왜도의 값은 양수나 음수가 될 수 있으며 정의되지 않을 수 있다.  
  
  
- 왜도 < 0
  - 확률밀도함수의 <font style='background-color:yellow; color: black'>왼쪽 부분에 긴 꼬리</font>를 가짐
  - 중앙값을 포함함 자료가 오른쪽에 더 많이 분포해 있음  
  
  
- 왜도 > 0
  - 확률밀도함수의 <font style='background-color:yellow; color: black'>오른쪽 부분에 긴 꼬리</font>를 가짐
  - 자료가 왼쪽에 더 많이 분포해 있음  
  
  
- 왜도 = 0
  - 평균과 중앙값이 같음

```python
# 개별 종목 수익률 왜도 값 구하기
df_norm.skew()
```

```python
# 참고) 개별 종목 수익률 히스토그램
h = df_norm.hist(figsize=(10,10))
```

<br>

### 5.7.2 첨도 (kurtosis)

<img src="./img/05_02.jpg" style="width:600px; margin-left: 10px;" />

- 확률분포의 뾰족한 정도를 나타내는 척도
- 관측치들이 어느 정도 집중적으로 중심에 몰려 있는 가를 측정할 때 사용  
  
  
- $K\text{(첨도값)} \approx 3$ : 산포도가 정규분포에 가까움
- $K\text{(첨도값)} \lt 3$ : 산포도가 정규분포보다 완만하게 납작한 분포
- $K\text{(첨도값)} \gt 3$ : 산포도가 정규분포보다 더 뾰족한 분포

```python
# 개별 종목 수익률 첨도 값 구하기
df_norm.kurtosis()
```

```python
# 참고) 개별 종목 수익률 히스토그램
h = df_norm.hist(figsize=(10,10))
```
